# Quick comparison of the old discoal with the new trajectory update

```python
#first simulate without selection
!./discoal 10 1000 10000 -r 20 -t 10 -en 0.05 0 0.1 | niceStats > test_stats
!./discoalTajUpdate 10 1000 10000 -r 20 -t 10 -en 0.05 0 0.1 | niceStats > testUpdate_stats
```

```python
#make boxplots in the seaborn style, do some install and imports
!conda install seaborn --yes
import pandas as pd
```

```python
df = pd.DataFrame()
import numpy as np
X1 = pd.read_table("test_stats")
X1=X1.assign(version=["old"]*1000)
X2 = pd.read_table("testUpdate_stats")
X2=X2.assign(version=["new"]*1000)
both = pd.concat((X1,X2))
df_long = pd.melt(both, "version", var_name="stat", value_name="value")
```

```python
import seaborn as sns
import matplotlib.pylab as plt
sns.factorplot("stat", hue="version", y="value", data=df_long, kind="box",size=8)
plt.show()
```

```python
# simulate with selection, but equilibrium
#!./discoal 10 1000 10000 -r 20 -t 100 -ws 0 -a 10000 | niceStats > test_stats
#!./discoalTajUpdate 10 1000 10000 -r 20 -t 100 -ws 0 -a 10000 | niceStats > testUpdate_stats
X1 = pd.read_table("test_stats")
X1=X1.assign(version=["old"]*1000)
X2 = pd.read_table("testUpdate_stats")
X2=X2.assign(version=["new"]*1000)
both = pd.concat((X1,X2))
df_long = pd.melt(both, "version", var_name="stat", value_name="value")
sns.factorplot("stat", hue="version", y="value", data=df_long, kind="box",size=8)
plt.ylim(-2.7,10)
plt.show()
```
